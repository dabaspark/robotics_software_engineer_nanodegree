```python
import matplotlib.pyplot as plt  # Import the pyplot module from matplotlib for plotting
import matplotlib.image as mpimg  # Import the image module from matplotlib for reading and displaying images

# Path to the image file
image_path = 'path_to_your_image.jpg'

# Read the image using matplotlib's imread function
image = mpimg.imread("sample.jpg")  # Read the image file and store it in the 'image' variable

# Display the image using matplotlib's imshow function
plt.imshow(image)  # Display the image using the imshow function

# Show the plot with the image
plt.show()  # Show the plot with the image
```

```python
import numpy as np

# Get the data type of the image
data_type = image.dtype

# Get the shape of the image
data_shape = image.shape

# Get the maximum and minimum values of the image
max_value = np.max(image)
min_value = np.min(image)

print("Data type:", data_type)
print("Data shape:", data_shape)
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Split the image into RGB channels
red_channel = image.copy()
red_channel[:, :, 1] = 0  # Zero out the green channel
red_channel[:, :, 2] = 0  # Zero out the blue channel

green_channel = image.copy()
green_channel[:, :, 0] = 0  # Zero out the red channel
green_channel[:, :, 2] = 0  # Zero out the blue channel

blue_channel = image.copy()
blue_channel[:, :, 0] = 0  # Zero out the red channel
blue_channel[:, :, 1] = 0  # Zero out the green channel

# Create a figure with three subplots to display the channels
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# Plot the red channel
axes[0].imshow(red_channel)
axes[0].set_title('Red Channel')

# Plot the green channel
axes[1].imshow(green_channel)
axes[1].set_title('Green Channel')

# Plot the blue channel
axes[2].imshow(blue_channel)
axes[2].set_title('Blue Channel')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np

def binary_image(image, rgb_thresh=(160, 160, 160)):
    # Create a binary image of the same shape as the input image
    binary = np.zeros_like(image[:,:,0])
    
    # Set pixels to 255 if they exceed the thresholds for each channel
    binary[(image[:, :, 0] > rgb_thresh[0]) & 
           (image[:, :, 1] > rgb_thresh[1]) &
           (image[:, :, 2] > rgb_thresh[2])] = 1
    
    return binary
```

```python
# Define the thresholds for each channel
red_threshold = 160
green_threshold = 160
blue_threshold = 160
rgb_thresh = (red_threshold, green_threshold, blue_threshold)

# Call the binary_image function
binary_img = binary_image(image, rgb_thresh)

# Display the image using matplotlib's imshow function
plt.imshow(binary_img*255, cmap='gray')

# Show the plot with the image
plt.show()
```

```python
# Display the original image and binary               
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(21, 7), sharey=True)
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=40)

ax2.imshow(binary_img, cmap='gray')
ax2.set_title('Your Result', fontsize=40)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
plt.show() 
```
